```python
import pandas as pd
import re

# Charger le fichier Excel
file_path = "mwe_occurrences_wide.xlsx"
df = pd.read_excel(file_path)

# Trouver la colonne la plus à droite contenant des occurrences
occ_columns = [col for col in df.columns if str(col).startswith("occ_")]
rightmost_col = occ_columns[-1]

# Extraire le niveau CECRL depuis les premiers crochets
niveau_cecrl = (
    df[rightmost_col]
    .astype(str)
    .str.extract(r"\[([^\]]+)\]")
)

# Ajouter la colonne
df["niveau_CECRL"] = niveau_cecrl

# Déplacer la colonne en 3e position
cols = df.columns.tolist()
cols.insert(2, cols.pop(cols.index("niveau_CECRL")))
df = df[cols]

# Sauvegarder le fichier
output_path = "mwe_occurrences_wide_with_cecrl.xlsx"
df.to_excel(output_path, index=False)

print(f"Fichier sauvegardé : {output_path}")
```
